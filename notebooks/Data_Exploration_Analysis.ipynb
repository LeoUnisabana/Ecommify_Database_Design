## 3️⃣ Análisis de Calidad de Datos y Valores Nulos

```python
def analizar_calidad_datos(df, nombre_dataset):
    """
    Función para analizar la calidad de datos de un DataFrame
    """
    print("=" * 80)
    print(f"ANÁLISIS DE CALIDAD: {nombre_dataset}")
    print("=" * 80)
    
    # Dimensiones
    print(f"\n📊 Dimensiones: {df.shape[0]:,} filas x {df.shape[1]} columnas")
    
    # Valores nulos
    print("\n🔍 Valores Nulos:")
    nulos = df.isnull().sum()
    porcentaje_nulos = (nulos / len(df)) * 100
    
    resumen_nulos = pd.DataFrame({
        'Columna': nulos.index,
        'Valores Nulos': nulos.values,
        'Porcentaje (%)': porcentaje_nulos.values
    })
    resumen_nulos = resumen_nulos[resumen_nulos['Valores Nulos'] > 0].sort_values('Valores Nulos', ascending=False)
    
    if len(resumen_nulos) > 0:
        display(resumen_nulos)
    else:
        print("✅ No hay valores nulos en este dataset")
    
    # Duplicados
    duplicados = df.duplicated().sum()
    print(f"\n🔄 Registros duplicados: {duplicados:,} ({(duplicados/len(df)*100):.2f}%)")
    
    # Tipos de datos
    print("\n📝 Tipos de datos:")
    print(df.dtypes.value_counts())
    
    print("\n" + "=" * 80 + "\n")

# Analizar cada dataset
analizar_calidad_datos(df_orders, "ORDERS")
analizar_calidad_datos(df_customers, "CUSTOMERS")
analizar_calidad_datos(df_order_items, "ORDER_ITEMS")
analizar_calidad_datos(df_products, "PRODUCTS")
analizar_calidad_datos(df_sellers, "SELLERS")
analizar_calidad_datos(df_payments, "PAYMENTS")
analizar_calidad_datos(df_reviews, "REVIEWS")
analizar_calidad_datos(df_geolocation, "GEOLOCATION")
```

### Visualización de Valores Nulos

```python
# Crear visualización de valores nulos para cada dataset
datasets = {
    'Orders': df_orders,
    'Customers': df_customers,
    'Order Items': df_order_items,
    'Products': df_products,
    'Sellers': df_sellers,
    'Payments': df_payments,
    'Reviews': df_reviews,
    'Geolocation': df_geolocation
}

fig, axes = plt.subplots(4, 2, figsize=(16, 20))
axes = axes.flatten()

for idx, (nombre, df) in enumerate(datasets.items()):
    nulos_pct = (df.isnull().sum() / len(df)) * 100
    nulos_pct = nulos_pct[nulos_pct > 0].sort_values(ascending=True)
    
    if len(nulos_pct) > 0:
        nulos_pct.plot(kind='barh', ax=axes[idx], color='coral')
        axes[idx].set_title(f'{nombre} - Valores Nulos (%)', fontsize=12, fontweight='bold')
        axes[idx].set_xlabel('Porcentaje (%)')
        axes[idx].grid(axis='x', alpha=0.3)
    else:
        axes[idx].text(0.5, 0.5, f'{nombre}\nSin valores nulos ✅', 
                      ha='center', va='center', fontsize=14, 
                      bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
        axes[idx].set_xticks([])
        axes[idx].set_yticks([])

plt.tight_layout()
plt.show()
```

## 4️⃣ Análisis Temporal

### 4.1 Preparación de datos temporales

```python
# Convertir columnas de fecha a datetime
date_columns = ['order_purchase_timestamp', 'order_approved_at', 
                'order_delivered_carrier_date', 'order_delivered_customer_date', 
                'order_estimated_delivery_date']

for col in date_columns:
    df_orders[col] = pd.to_datetime(df_orders[col])

# Extraer componentes de fecha
df_orders['year'] = df_orders['order_purchase_timestamp'].dt.year
df_orders['month'] = df_orders['order_purchase_timestamp'].dt.month
df_orders['month_name'] = df_orders['order_purchase_timestamp'].dt.month_name()
df_orders['day'] = df_orders['order_purchase_timestamp'].dt.day
df_orders['dayofweek'] = df_orders['order_purchase_timestamp'].dt.dayofweek
df_orders['dayofweek_name'] = df_orders['order_purchase_timestamp'].dt.day_name()
df_orders['quarter'] = df_orders['order_purchase_timestamp'].dt.quarter

print("✅ Datos temporales procesados")
print(f"\nRango de fechas: {df_orders['order_purchase_timestamp'].min()} a {df_orders['order_purchase_timestamp'].max()}")
print(f"Total de días: {(df_orders['order_purchase_timestamp'].max() - df_orders['order_purchase_timestamp'].min()).days}")
```

### 4.2 Evolución de Pedidos en el Tiempo

```python
# Agrupar por mes-año
df_orders['year_month'] = df_orders['order_purchase_timestamp'].dt.to_period('M')
pedidos_mes = df_orders.groupby('year_month').size().reset_index(name='num_pedidos')
pedidos_mes['year_month'] = pedidos_mes['year_month'].astype(str)

# Visualización con plotly
fig = px.line(pedidos_mes, x='year_month', y='num_pedidos',
              title='Evolución de Pedidos por Mes (2016-2018)',
              labels={'year_month': 'Mes', 'num_pedidos': 'Número de Pedidos'},
              markers=True)

fig.update_layout(hovermode='x unified', height=500)
fig.show()

# Estadísticas por año
print("\n📊 Pedidos por año:")
print(df_orders.groupby('year').size().to_frame('Total Pedidos'))
```

### 4.3 Estacionalidad - Día de la Semana

```python
# Pedidos por día de la semana
pedidos_dia = df_orders['dayofweek_name'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

fig = px.bar(x=pedidos_dia.index, y=pedidos_dia.values,
             title='Distribución de Pedidos por Día de la Semana',
             labels={'x': 'Día de la Semana', 'y': 'Número de Pedidos'},
             color=pedidos_dia.values,
             color_continuous_scale='Blues')

fig.update_layout(showlegend=False, height=500)
fig.show()
```

### 4.4 Estacionalidad - Mes del Año

### 4.5 Análisis de Tiempos de Entrega

```python
# Calcular diferencias de tiempo
df_orders['delivery_time'] = (df_orders['order_delivered_customer_date'] - 
                               df_orders['order_purchase_timestamp']).dt.days

df_orders['estimated_delivery_time'] = (df_orders['order_estimated_delivery_date'] - 
                                         df_orders['order_purchase_timestamp']).dt.days

df_orders['delivery_delay'] = df_orders['delivery_time'] - df_orders['estimated_delivery_time']

# Filtrar solo pedidos entregados
delivered_orders = df_orders[df_orders['order_status'] == 'delivered'].copy()

print("=" * 80)
print("ANÁLISIS DE TIEMPOS DE ENTREGA")
print("=" * 80)
print(f"\n📦 Pedidos entregados: {len(delivered_orders):,}")
print(f"\n⏱️ Tiempo promedio de entrega: {delivered_orders['delivery_time'].mean():.1f} días")
print(f"⏱️ Mediana de tiempo de entrega: {delivered_orders['delivery_time'].median():.1f} días")
print(f"\n⚠️ Promedio de retraso: {delivered_orders['delivery_delay'].mean():.1f} días")
print(f"✅ Pedidos entregados a tiempo o antes: {(delivered_orders['delivery_delay'] <= 0).sum():,} ({(delivered_orders['delivery_delay'] <= 0).sum()/len(delivered_orders)*100:.1f}%)")
print(f"⏰ Pedidos entregados con retraso: {(delivered_orders['delivery_delay'] > 0).sum():,} ({(delivered_orders['delivery_delay'] > 0).sum()/len(delivered_orders)*100:.1f}%)")

# Visualización
fig = go.Figure()

fig.add_trace(go.Histogram(
    x=delivered_orders['delivery_time'],
    name='Tiempo Real de Entrega',
    opacity=0.7,
    marker_color='steelblue'
))

fig.add_trace(go.Histogram(
    x=delivered_orders['estimated_delivery_time'],
    name='Tiempo Estimado de Entrega',
    opacity=0.7,
    marker_color='orange'
))

fig.update_layout(
    title='Distribución de Tiempos de Entrega (Real vs Estimado)',
    xaxis_title='Días',
    yaxis_title='Frecuencia',
    barmode='overlay',
    height=500
)

fig.show()
```

```python
# Merge para obtener categorías traducidas
df_products_translated = df_products.merge(
    df_category_translation, 
    on='product_category_name', 
    how='left'
)

# Merge con order_items para ver qué productos se venden más
df_items_products = df_order_items.merge(
    df_products_translated[['product_id', 'product_category_name_english']], 
    on='product_id', 
    how='left'
)

# Top categorías
top_categories = df_items_products['product_category_name_english'].value_counts().head(20)

fig = px.bar(y=top_categories.index, x=top_categories.values,
             title='Top 20 Categorías de Productos Más Vendidas',
             labels={'y': 'Categoría', 'x': 'Número de Ventas'},
             orientation='h',
             color=top_categories.values,
             color_continuous_scale='Viridis')

fig.update_layout(showlegend=False, height=600, yaxis={'categoryorder':'total ascending'})
fig.show()

print("\n📦 Estadísticas de Productos:")
print(f"Total de productos únicos: {df_products['product_id'].nunique():,}")
print(f"Total de categorías: {df_products['product_category_name'].nunique()}")
print(f"Productos sin categoría: {df_products['product_category_name'].isna().sum()}")
```

### 6.2 Análisis de Dimensiones de Productos

```python
# Estadísticas de dimensiones
print("=" * 80)
print("ANÁLISIS DE DIMENSIONES DE PRODUCTOS")
print("=" * 80)

dimension_cols = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

print("\n📏 Estadísticas de dimensiones:")
display(df_products[dimension_cols].describe())

# Visualización de distribuciones
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for idx, col in enumerate(dimension_cols):
    # Quitar outliers extremos para mejor visualización
    data = df_products[col].dropna()
    q99 = data.quantile(0.99)
    data_filtered = data[data <= q99]
    
    axes[idx].hist(data_filtered, bins=50, color='skyblue', edgecolor='black')
    axes[idx].set_title(f'Distribución: {col}', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frecuencia')
    axes[idx].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
